```python
with open('chaves-tt-app.txt', 'r') as tfile:
    consumer_key = tfile.readline().strip('\n')
    consumer_secret = tfile.readline().strip('\n')
    access_token = tfile.readline().strip('\n')
    access_token_secret = tfile.readline().strip('\n')
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tw.API(auth)

public_tweets = api.home_timeline()
```

```python
qtd_tweets = 50
query_search = "#covid" + " -filter:retweets"
cursor_tweets = tw.Cursor(api.search, 
                   q=query_search).items(qtd_tweets)
```

```python
#stop_words = set(stopwords.words('english'))
i = 0
for tweet in cursor_tweets:
    #removendo stopwords
    sentenca = tweet._json['text']
    
    #word_tokens = word_tokenize(sentenca)
    #filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    #filtered_sentence = []
    #for w in word_tokens:
    #    if w not in stop_words:
    #        filtered_sentence.append(w)
    
    #Adicionando tweets no dicion√°rio
    disc_tweets[i] = sentenca
    i += 1
    print(sentenca)
    print("\n__________________________________\n")
```

```python
print(disc_tweets)
```

```python
#tratando os dados
characters = "?.!/\*#$¬£%¬®&,|:;¬∞¬∫¬™¬ß)("

for i in disc_tweets:
    texto_tweet = disc_tweets[i]
    
    #removendo URL
    texto_tweet = re.sub('http://\S+|https://\S+', '', texto_tweet)
    
    #removendo marca√ß√µes de usu√°rios @
    texto_tweet = re.sub("(@[A-Za-z0-9]+)","",texto_tweet)
    
    #removendo caracteres e pula linha
    for x in range(len(characters)):
        texto_tweet = texto_tweet.replace(characters[x],"")
        texto_tweet = texto_tweet.replace('\n'," ")
    
    
    disc_tweets[i] = texto_tweet
```

```python
i = 1
for i in range(len(disc_tweets)):
    print(disc_tweets[i]+"\n")
    i+=1
```
