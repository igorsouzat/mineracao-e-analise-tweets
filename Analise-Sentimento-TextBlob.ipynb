```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import csv
```

```python
count_pos = 0
count_neg = 0
p_pos = 0
p_neg = 0
disc_resultados = {0: [None, None, None, None, None]}
```

```python
i = 0
with open('frases.csv', 'r', encoding="utf8") as frases:
    reader = csv.reader(frases)
    
    for linha in reader:
        frase = linha[1]
        blob = TextBlob(frase)
            
        #traduzindo frase para o inglês
        if blob.detect_language() != 'en':    
            frase_traduzida = blob.translate(to='en')
            
            #realizando a análise de sentimento
            blob = TextBlob(frase_traduzida.raw, analyzer=NaiveBayesAnalyzer())
                        
            p_pos = blob.sentiment[1]*100
            p_neg = blob.sentiment[2]*100
            
            print(frase_traduzida)
            print("Positivo: {}\nNegativo: {}\n".format(p_pos, p_neg))
            
            disc_resultados[i] = [frase, frase_traduzida, blob.sentiment[0],p_pos, p_neg]
        else:
            blob = TextBlob(frase, analyzer=NaiveBayesAnalyzer())
            
            p_pos = blob.sentiment[1]*100
            p_neg = blob.sentiment[2]*100
            
            print(frase)
            print("Positivo: {}\nNegativo: {}\n".format(p_pos, p_neg))
            
            disc_resultados[i] = [frase, frase, blob.sentiment[0], p_pos, p_neg]
            
        if blob.sentiment[0] == "pos":
            count_pos+=1
        else:
            count_neg+=1
            
        i += 1
        
```

```python
print("Positivos: {}\nNegativos: {}".format(count_pos, count_neg))
```

```python
for i in disc_resultados:
    if disc_resultados[i][3] == disc_resultados[i][4]:
        print("Orignal: {}\nInglês: {}\nSentimento: Neutro\nPositivo: {}\nNegativo: {}\n\n".format(disc_resultados[i][0], disc_resultados[i][1], disc_resultados[i][3], disc_resultados[i][4]))
    else:
        print("Orignal: {}\nInglês: {}\nSentimento: {}\nPositivo: {}\nNegativo: {}\n\n".format(disc_resultados[i][0], disc_resultados[i][1], disc_resultados[i][2], disc_resultados[i][3], disc_resultados[i][4]))
```
